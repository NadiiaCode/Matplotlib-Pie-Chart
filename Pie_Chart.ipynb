```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# read CSV file
gdp_by_country = pd.read_csv("data.csv")
# print data from CSV file
print(gdp_by_country)
```

```python
country = gdp_by_country["Country"]
estimate = gdp_by_country["Estimate"]
plt.pie(estimate, labels=country)
plt.title("GDP by country (without USA and China)")
plt.show()
```

```python
# create simple pie chart
country = gdp_by_country["Country"]
estimate = gdp_by_country["Estimate"]
plt.pie(estimate, labels=country)
# add title
plt.title("GDP by country (without USA and China)")
plt.show()
```

```python
country = gdp_by_country["Country"]
estimate = gdp_by_country["Estimate"]
# create list of colors (total 10)
# add edgecolor for better visibility
colors = ["#9A6324", "#e6194B", "#f58231", "#ffe119", "#3cb44b", "#42d4f4",
          "#4363d8", "#911eb4", "#f032e6", "#a9a9a9"]
plt.pie(estimate, labels=country, colors=colors, wedgeprops={"edgecolor": "#444444"})
plt.title("GDP by country (without USA and China)")
plt.show()
```

```python
country = gdp_by_country["Country"]
estimate = gdp_by_country["Estimate"]
colors = ["#9A6324", "#e6194B", "#f58231", "#ffe119", "#3cb44b", "#42d4f4",
          "#4363d8", "#911eb4", "#f032e6", "#a9a9a9"]
# add autopct for % of each country in pie
# add shadow 
# change edgecolor to black
plt.pie(estimate, labels=country, colors=colors,
        autopct='%1.1f%%', shadow=True, wedgeprops={"edgecolor": "#000000"})
plt.title("GDP by country (without USA and China)")
plt.show()
```

```python
country = gdp_by_country["Country"]
estimate = gdp_by_country["Estimate"]
colors = ["#9A6324", "#e6194B", "#f58231", "#ffe119", "#3cb44b", "#42d4f4",
          "#4363d8", "#911eb4", "#f032e6", "#a9a9a9"]
# add explode to move one or more part of pie away from the main pie
explode = [0.15, 0.15, 0, 0, 0, 0, 0, 0, 0, 0]
# add startangle for pie rotation 
# add pctdistance for changing the distance of displayed % from the center
# add labeldistance to display label farther (closer) to the pie
plt.pie(estimate, labels=country, colors=colors, explode=explode,
        autopct='%1.1f%%', shadow=True, startangle=90, wedgeprops={"edgecolor": "#000000"},
        pctdistance=0.8,  labeldistance=1.2)
plt.title("GDP by country (without USA and China)")
plt.show()
```

```python
country = gdp_by_country["Country"]
estimate = gdp_by_country["Estimate"]
colors = ["#9A6324", "#e6194B", "#f58231", "#ffe119", "#3cb44b", "#42d4f4",
          "#4363d8", "#911eb4", "#f032e6", "#a9a9a9"]
explode = [0.15, 0.15, 0, 0, 0, 0, 0, 0, 0, 0]
# remove shadow
# add linewidth (higher number = thicker line) 
# add text color and size 
# add radius to increase the size of the pie
plt.pie(estimate, labels=country, colors=colors, explode=explode,
        autopct='%1.1f%%', startangle=90, 
        wedgeprops={"edgecolor": "#000000", "linewidth": 2}, 
        textprops = dict(color ="#444400", size=15),
        pctdistance=0.8,  labeldistance=1.2, radius=1.4)
# add "y" for moving title away from the pie
# add text size, color and weight
plt.title("GDP by country (without USA and China)", y=1.3, size=20, color="#444400", weight="bold")
plt.show()
```

```python
country = gdp_by_country["Country"]
estimate = gdp_by_country["Estimate"]
colors = ["#9A6324", "#e6194B", "#f58231", "#ffe119", "#3cb44b", "#42d4f4",
          "#4363d8", "#911eb4", "#f032e6", "#a9a9a9"]
explode = [0.15, 0.15, 0, 0, 0, 0, 0, 0, 0, 0]
# add font weight bold
# add rotatelabels (360 deg) to display countries names (labels) around the pie
plt.pie(estimate, labels=country, colors=colors, explode=explode,
        autopct='%1.1f%%', startangle=90, 
        wedgeprops={"edgecolor": "#000000", "linewidth": 2},
        textprops = dict(color ="#444400", size=15, weight="bold"), pctdistance=0.8, 
        labeldistance=1.1, radius=1.4, rotatelabels = 360)
plt.title("GDP by country (without USA and China)", y=1.4, size=20, color="#444400", weight="bold")
# save pie chart as png file
plt. savefig("1.png")
plt.show()
```
